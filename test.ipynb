```python
import os
vehicle_models_master_df = []
for file in os.listdir('us-car-models-data'):
  f = os.path.join('us-car-models-data', file)
  if f.endswith('.csv'):
  
    vehicle_models_master_df.append(pd.read_csv(f))
vehicle_models_master_df = pd.concat(vehicle_models_master_df, ignore_index=True)

models_df = vehicle_models_master_df
models_df
```

```python
def get_model(year, make):
  modelsYear = models_df.loc[models_df["year"] == year]
  makeModels = modelsYear.loc[modelsYear["make"] == make]
  return makeModels



i = 0
for make in range(len(vehicle_df[:100])):
  year = vehicle_df['year'][i]
  make = vehicle_df["make"][i]
  models = get_model(int(year), make).model
  model = [word for word in models if word in vehicle_df.title[i]]
  model = ''.join(model)
  vehicle_df["model"][i] = model
  i += 1
  
vehicle_df.head(40)
```
